```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\Dell\Desktop\Customer Call List.xlsx")
df
```

```python
#drop not useful cloumns
df.drop(columns="Not_Useful_Column")
```

```python
#Cleaning column data
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
```

```python
df["Last_Name"] 
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"[\/|\-]", "", regex=True)
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:9])
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--", "")
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--", "")
```

```python
df
```

```python
df[["street_address","State","Zip_code"]]=df["Address"].str.split(",",n=2,expand=True)
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df=df.replace('N/a','')
##df=df.replace('NaN','')
df=df.fillna('')
```

```python
df
```

```python
#Removing colum if it is is Not required to contact 
#.loc is used to select data using row labels and column names
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=='Y':
        df.drop(x, inplace=True)

        
```

```python
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"]=='':
        df.drop(x, inplace=True)


#alos another way to drop null values
#df=df.dropna(subset="Phone_number"),inplace=True)
```
